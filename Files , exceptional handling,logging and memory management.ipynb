7. How can you handle multiple exceptions in python ?

 ▶ In python, multiple eceptions can be handled using a single try block followed by multiple except block can specify a diiferent exception type to catch.

Here's how to do it:

1. Using multiple except blocks:
This is the most common handle different exceptions separately. Each except block handles a specific exception type.

```python
try:
  # code that might raise an exception
  result=10/int(input("Enter a number: "))
  print(result)
except ValueError:
  print("invalid input. Please enter a valid integer.")
except ZeroDivisionError:
  print("cannot divide by zero")
except Exception as e:
  print("an unexpected error occured:",e)
```

 2. Using a single except block with a tuple of exceptions:
 If you want to handle multiple exceptions in the same way, you can specify them as a tuple within a single except block.  

```python
try:
  # code that might raise an exception
  result=10/int(input("Enter a number: "))
  print(result)
except (ValueError,ZeroDivisionError):
  print("invalid input or cannot be divided by zero")
except Exception as e:
  print("an unexpected error occured:",e)
```

 3. Using a base exception class:
 You can also catch a group of exceptions by specifying a base class that they inherit from. For example, OSError is a base class for file-related exceptions.

```python
try:
  # code that raise file-related exceptions
  with open("nonexistent_file.txt","r") as f:
    content=f.read()
except OSError:
  print("file operation failed")
except Exception as e:
  print("an unexpected error occured:",e)
```

 4. Using nested try-excepted blocks:
 You can nest try-except blocks to handle exceptions in different parts of your code.


```python
try:
  # outer try block
  try:
    # inner try block
     result=10/int(input("Enter a number: "))
     print(result)
  except ValueError:
   print("invalid input. Please enter a valid integer.")
except ZeroDivisionError:
  print("cannot divide by zero")
except Exception as e:
  print("an unexpected error occured:",e)
```

 Use cases:- Sutable for CPU-bound tasks,such as complex calculations, data processing and scientific simulations.

12. What are the basic steps involved in exception handling in Python ?

 ▶ Steps involved in exception handling in python:

 try block:This block encloses the code that might raise an exception. Python will attempt to execute the code within the try block.

 except block:If an excpetion occurs within try block, python immediately jumps to the execpt block. The except block specifies how to handle the exception . You can have multiple except blocks to handle different types of exceptions.

 else block(optional):The else block is executed only if no exception is raised in the try block. It is useful for code that should run if the try block succeeds without errors.

 finally block(optional):The finally block is always executed, reagardless of whether an exception occured or not. It's typically used for cleanup actions, such as closing files or releasing resources.  



```python
try:
  result=10/int(input("Enter a number: "))
  print(result)
except ValueError:
  print("invalid input. Please enter a valid integer.")
except ZeroDivisionError:
  print("cannot divide by zero")
else:
  print("Result :",result)
finally:
  print("Execution completed")
```

18. What is the difference between os.fork() and multiprocessing in Python ?

 ▶ os.fork()

  1. low level system call:- os.fork() is a direct interface to the operating system's fork system call. It creates a new process that is an exact copy of calling process.

  2. copy on write:- It uses copy-on-write mechanism, which means that child process intially shares the parent's memory. Changes made by either process are only copied when necessary.

  3. Unix-like systems:- os.fork() is pramarily available on unix-like systems(linux,macOS).It's not supported on windows.

  4. Potential issues:- When used with multithread programs , os.fork() can lead to issues like deadlocks because only the calling thread is duplicated in the child process.Locks held by other threads in the parent process won't be released in the child process.

  5. Memory usage:- Forking lead to increased memory usage if both parent and child processes start modiying the shared memory, causing it to be copied.

  ▶ Multiprocessing module:-

  1. High level abstration:The multiprocessing module provides a higher-level interface for creating and managing processes.

  2. cross-platform: It works on both unix-like systems and windows.

  3. process manangement: It provides tools for inter-process communication(IPC),such as queues and pipe, which allow processes to exchange data.

  4. Better for multithread programs: Multiprocessing is generally safer to use with multithread programs, especially when using the spawn start method.

  5. More robust: It handles process creation and management more robustly than using os.fork() directly.

23. What are the challenges associated with memory management in Python ?

 ▶ Challenges with Python memory management:

  1. Unpredictable Usage: Automatic management (reference counting, garbage collection) makes precise memory control difficult and can cause performance pauses during garbage collection cycles.
  2. Memory Leaks: Despite automatic GC, leaks can occur due to:
   * Circular references: Objects referencing each other prevent deallocation.
  * Unreleased external resources: Forgetting to close file handles, network connections, etc.
  * Long-lived objects/globals: Objects held indefinitely can accumulate memory.
   3. GIL & Concurrency: The Global Interpreter Lock (GIL) limits true parallelism in multi-threaded Python, impacting memory-intensive parallel tasks.
  4. Debugging Difficulty: Pinpointing memory leaks or inefficiencies can be challenging due to Python's abstraction of memory internals.
  5. Large Data Efficiency: In-memory processing of large datasets can lead to high memory usage and "out of memory" errors, exacerbated by object overhead and immutable type operations creating new objects.

 18. How would you implement a basic logging setup that logs to a file with rotation after 1MB ?

```python
import logging
from logging.handlers import RotatingFileHandler

# Set up logger
logger = logging.getLogger("MyLogger")
logger.setLevel(logging.DEBUG)  # Set the minimum log level

# Create a rotating file handler (rotate after 1MB, keep 3 backups)
handler = RotatingFileHandler("app.log", maxBytes=1 * 1024 * 1024, backupCount=3)
handler.setLevel(logging.DEBUG)

# Define a log format
formatter = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s')
handler.setFormatter(formatter)

# Add the handler to the logger
logger.addHandler(handler)

# Example usage
for i in range(10):
    logger.debug(f"This is log message number {i}")
```

21.  Write a Python program that reads a file and prints the number of occurrences of a specific word.

```python
def count_word_occurrences(filename, target_word):
    try:
        with open(filename, 'r') as file:
            content = file.read().lower()  # Read and convert to lowercase
            word_list = content.split()   # Split into words
            count = word_list.count(target_word.lower())  # Count target word
            print(f"The word '{target_word}' occurred {count} times in '{filename}'.")
    except FileNotFoundError:
        print(f"Error: The file '{filename}' does not exist.")

# Example usage
file_name = "sample.txt"
word_to_search = "python"
count_word_occurrences(file_name, word_to_search)
```
